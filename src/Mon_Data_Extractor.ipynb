```python
para = ""
for topic in topics:
  result = wikipedia.search(topic)
  print(result)
  content_orig = wikipedia.page(result[0]).content

  content = content_orig.split("နိဿဲ")[0]
  content = content.split("ဆက်ဗှ်")[0]
  content = content.split("သရာဲသ္ဂံင်ဂၠိုၚ်")[0]
  content = content.split("စရင်ပြကိုဟ် မနိဿဲဂမၠိုင်")[0]
  content = content.split("အယုတ်ဍုၚ်")[0]
  content = content.split("လေန်မ္ၚး")[0]
  content = content.split("လေန် မ္ၚးဂမၠိင်")[0]
  st = re.sub(r"==.*?==+", "", content)
  st = re.sub("(/|\?|\n|\t)", "", st)
  para += st + "၊၊"
```
