```python
para = ""
for topic in topics:
  result = wikipedia.search(topic)
  print(result)
  content_orig = wikipedia.page(result[0]).content

  content = content_orig.split("ရည်ညွှန်းချက်များ")[0]
  content = content.split("ကိုးကား")[0]
  content = content.split("မှတ်စု")[0]
  content = content.split("ဆက်စပ်ကြည့်ရှုရန်")[0]
  st = re.sub(r"==.*?==+", "", content)
  st = re.sub("(/|\?|\n|\t)", "", st)
  para += st + "။"
```

```python
data = para.split("။")
```
